```python
import pystan
import arviz
```

# this code is for bin prob experiencing

```python
coin_code = """
data {
    int<lower=0> n; // number of tosses
    int<lower=0> y; // number of heads
}
transformed data {}
parameters {
    real<lower=0, upper=1> p;
}
transformed parameters {}
model {
    p ~ beta(2, 2);
    y ~ binomial(n, p);
}
generated quantities {}
"""

coin_dat = {
             'n': 100,
             'y': 61,
            }
```

```python
sm = pystan.StanModel(model_code=coin_code)
```

```python
fit = sm.sampling(data=coin_dat, iter=1000, chains=4, control = {"adapt_delta" : 0.9})
```

```python
plt = fit.plot()
```
